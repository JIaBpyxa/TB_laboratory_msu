```python
months = data_set.transpose()[0]
months
```

```python
deltas = data_set.transpose()[1]
deltas
```

```python
def Lambda_1_n(x):
    answer = 0
    for j in range(set_length):
        answer += (deltas[j] * int(months[j] <= x)) / (np.sum(months >= months[j]))
    return answer
```

```python
def Lambda_n(x):
    answer = 0
    for j in range(set_length):
        answer += (int(months[j] <= x)) / (np.sum(months >= months[j]))
    return answer
```

```python
def H_n(x):
    return np.sum(months < x) / set_length

# 1 - H_n(u -)
def First_H(u):
    answer = 0
    for j in range(set_length):
        answer += (int(months[j] >= u))
    answer = answer / set_length
    return answer

# 1 - H_n(u)
def Second_H(x):
    answer = 0
    for j in range(set_length):
        answer += (int(months[j] > x))
    answer /= set_length
    return answer

# H_1n(u) - H_1n(u-)
def Third_H(u):
    answer = 0
    for j in range(set_length):
        answer += deltas[j] * (int(months[j] == u))
    answer /= set_length
    return answer
```

```python
def F_AB_n(x):
    if x < np.min(months):
        return 0
    if x > np.max(months):
        return 1
    #answer = 0
    #for u in range(x + 1):
    #   answer += Third_H(u) / First_H(u)
    #answer = 1 - np.exp(-1 * answer)
    return 1 - np.exp(-1 * Lambda_1_n(x))
```

```python
def F_KM_n(x):
    if x < np.min(months):
        return 0
    if x > np.max(months):
        return 1
    answer = 1
    for u in range(x + 1):
        if First_H(u) == 0:
            print(u)
        answer *= 1 - (Third_H(u) / First_H(u))
    answer = 1 - answer
    return answer
```

```python
def F_RR_n(x):
    if x < np.min(months):
        return 0
    if x > np.max(months):
        return 1
    answer = 1 - np.power(Second_H(x), (Lambda_1_n(x) / Lambda_n(x)))
    return answer
```

```python
def F_ACL_n(x):
    if x < np.min(months):
        return 0
    if x > np.max(months):
        return 1
    return 1 - (np.power((1 - H_n(x)), (np.sum(deltas) / set_length)))
```

Доверительная полоса

```python
def d_n(x):
    answer = 0
    for j in range(set_length):
        answer += (deltas[j] * int(months[j] <= x)) / np.power((np.sum(months >= months[j])), 2)
    answer *= set_length
    return answer
```

```python
T = 1153

def M_minus_n(x, F):
    return F - np.power(set_length, -0.5) * (1 - F) * (np.sqrt(d_n(T)) + 1.37 * d_n(x) / np.sqrt(d_n(T)))

def M_plus_n(x, F):
    return (F + np.power(set_length, -0.5) * (np.sqrt(d_n(T)) + 1.37 * d_n(x) / np.sqrt(d_n(T)))) / (1 + np.power(set_length, -0.5) * (np.sqrt(d_n(T)) + 1.37 * d_n(x) / np.sqrt(d_n(T))))
```
