```python
import numpy as np
```

```python
from numpy.random import default_rng

rng = default_rng(seed=1)
```

```python
set_length = 100
random_set = rng.integers(0, 100, set_length)
random_set
```



---


Выборка {X_1, ..., X_100}

```python
input_selection = random_set / set_length
input_selection = np.sort(input_selection)
input_selection
```



---
Выборка {Y_1, ..., Y_100}



```python
my_number = 12
selection = -my_number * np.log(1 - input_selection)
selection
```

```python
unique_selection, selection_count = np.unique(selection, return_counts=True)
print(unique_selection)
print(selection_count)
```



---
**Функция эмпирического распределения**


```python
def empirical_distribute(t):
    indexes, = np.where(unique_selection < t)
    counts = selection_count[indexes]
    return np.sum(counts) / set_length
```

```python
empirical_distribute(5)
```

---

**Ядерная оценка**

```python
from scipy.stats import norm
```

```python
def kernel_estimation(t):
    h = set_length ** (-1 / 3)
    result = 0

    for element, element_count in zip(unique_selection, selection_count):
        result += element_count * norm.cdf((t - element) / h)

    return result / set_length
```

```python
def distribution_function(t):
    return 1-np.exp(-t/my_number)
```

---

**Построение графика**

```python
distribution_function_y = distribution_function(unique_selection)
empirical_distribute_y = np.zeros(0)
kernel_estimation_y = np.zeros(0)

for t in unique_selection:
    empirical_distribute_y = np.append(empirical_distribute_y, empirical_distribute(t))
    kernel_estimation_y = np.append(kernel_estimation_y, kernel_estimation(t))

print(distribution_function_y)
print(empirical_distribute_y)
print(kernel_estimation_y)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(20,12))
plt.grid()
plt.plot(unique_selection, distribution_function_y, color = 'y', label = 'Distribution function y')
plt.plot(unique_selection, empirical_distribute_y, color = 'm', label = 'Empirical distribution')
plt.plot(unique_selection, kernel_estimation_y, color = 'c', label = 'Kernel estimation')
plt.legend()
plt.show()
```

---
**Доверительная полоса**

```python
significance_level = 0.05
interval = 1.36 / np.sqrt(set_length)

zeros = np.zeros(empirical_distribute_y.shape)
ones = np.ones(empirical_distribute_y.shape)

left_board = np.maximum(zeros, empirical_distribute_y - interval)
right_board = np.minimum(ones, empirical_distribute_y + interval)
```

```python
plt.figure(figsize=(20,12))
plt.grid()
plt.plot(unique_selection, empirical_distribute_y, color = 'c')
plt.fill_between(unique_selection, left_board, right_board, color = 'red', alpha = .5)
plt.show()
```

```python
def density_function(t):
    return 1/my_number * np.exp(-t/my_number)

def kernel_density_estimation(t):
    h = set_length ** (-1 / 3)
    result = 0

    for element, element_count in zip(unique_selection, selection_count):
        result += element_count * norm.pdf((t - element) / h)

    return result / (set_length * h)
```

```python
kernel_density_estimation_y = np.zeros(0)
arange = np.arange(0,100,0.1)
density_y = density_function(arange)
for t in arange:
    kernel_density_estimation_y = np.append(kernel_density_estimation_y, kernel_density_estimation(t))
```

```python
plt.figure(figsize=(20,12))
plt.grid()
plt.plot(arange, density_y, color = 'y', label = 'Density function y')
plt.plot(arange, kernel_density_estimation_y, color = 'c', label = 'Kernel density estimation')
plt.legend()
plt.show()
```
